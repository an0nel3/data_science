```python
%matplotlib inline
import csv as csv
import numpy as np
import time,timeit
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
from pandas import ewma
```

```python
repairs_df = pd.read_csv('RepairTrain.csv')
repairs = repairs_df[['module_category','component_category','year/month(repair)','number_repair']]
repairs.columns =['mods','comp','ymr','repair']
repairs['ym'] = pd.to_datetime(repairs.ymr)
df = repairs.groupby(['mods','comp','ym']).sum()
df = df.reset_index()
mappings = pd.read_csv('Output_TargetID_Mapping.csv')
mappings['ym'] = pd.to_datetime(mappings.year.astype(str) +"/" +mappings.month.astype(str))
mappings.columns=['mods','comp','year','month','ym']
maps = mappings[['mods','comp','ym']]
```

```python
def plot_all(df,maps,spanner=3):
    #look through mappings
    predictions=[]
    big_frame = pd.concat([df,maps])
    big_sorted = big_frame.sort(['mods','comp','ym'])
        
    repair_vals  = big_sorted.repair.values
    
    repair_size  = repair_vals.size
    
    
    for i in xrange(repair_vals.size) :
        if np.isnan(repair_vals[i]):
            
            one  = float(repair_vals[i-1])
            two = float(repair_vals[i-2])
            three  = float(repair_vals[i-3])
            four = float(repair_vals[i-4])
            
            first = (four-three)/four
            second = (three-two)/three
            third = (two-one)/3
            AVEG = (one+two+three)/3
            #for 3/20 sub 2, took the avg of last three 
            #repair_vals[i]=.8*AVEG
            repair_vals[i]=.9*one  
    
    for each_mod in maps.mods.unique():
        #look at each each component we need to check for
        for each_comp in maps[maps.mods==each_mod].comp.unique():
            data_plot = df[(df.mods ==each_mod) & (df.comp==each_comp)]
                           
            h = big_sorted[(big_sorted.mods ==each_mod) & (big_sorted.comp ==each_comp)]
            
            
            h_new = h[['ym','repair']]
            x=maps[(maps.mods ==each_mod) & (maps.comp==each_comp)].shape[0]
            if df[(df.mods==each_mod) & (df.comp==each_comp)].shape[0] > 3:
                ######span setter
                
                pred_holder = pd.DataFrame(pd.ewma(h_new.repair,spanner,adjust=False))   
                
                #####
                pred_holder['ym'] = h.ym
                pred_holder.columns = [['repair','ym']]
                preds_map = pred_holder.ix[maps[(maps.mods ==each_mod) & (maps.comp==each_comp)].index.values]
                preds = pred_holder.ix[maps[(maps.mods ==each_mod) & (maps.comp==each_comp)].index.values].repair
                #print preds
                
                if preds.size > x:
                    #print "problem %s,%s" %(each_mod, each_comp)
                    #print "problem for manys"
                    #print preds
                    preds=np.ones(x)
                predictions +=list(preds[:x])
                
            else:
                #for now, basically set to 0 since maps.repair == 0
                preds = h_new.ix[maps[(maps.mods ==each_mod) & (maps.comp==each_comp)].index.values].repair
                predictions+=list(preds[:x])
                if preds.size > x:
                    print "problem %s,%s" %(each_mod, each_comp)
                    print "problem is for short list"
                    #print preds
            pred_holder.plot(x='ym',title="%s, and %s" %(each_mod,each_comp))
            h_new.plot(x='ym',y='repair',style='ro')
    return np.array(predictions)                             
x2 = plot_all(df,maps,spanner=1)                                         
    
```
